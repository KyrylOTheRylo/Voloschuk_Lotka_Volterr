```python
import numpy as np
from matplotlib import pyplot
```

```python
n = 5000
m = 1000
Mm = 2500
Nmin = 5000
```

```python
def plot(t,a ,b ) -> (np.array, np.array) :
    """
    returns:
    """
    nt =[n]
    mt = [m]
    min1 = n
    for i in range(len(t)-1):
        if nt[i-1] < min1:
            min1 = nt[i-1]
        nt.append(nt[i] + a*nt[i]*(1 - mt[i]/Mm))
        mt.append(mt[i] + b*mt[i]*(nt[i]/Nmin - 1))
    print(min1)
    return nt, mt
```

```python
fig, ax = pyplot.subplots()
t = np.linspace(1, 500, 500 , dtype=int)
t = np.concatenate((t,t+500))
victim, predator = plot(t, 0.055, 0.035)
ax.plot(t , victim, label = "victim")
ax.plot(t , predator, label = 'predator')
ax.legend(loc='upper center')
pyplot.show()
```

```python
fig, ax = pyplot.subplots()
ax.scatter(t , predator, label = "predator", s =1 , color ="orange")
ax.scatter(t , victim, label = "victim", s =1, color = 'blue')
ax.legend(loc='upper center')
pyplot.show()
```
